Q5. Explain the concept of levels of measurement and give an example of a variable for each level.

Ans. Levels of measurement tell you how precisely variables are recorded.
        types : a] Nominal scale data : the data can only be categorized
                    eg: gender 
                b] Ordinal scale data : the data can be categorized and ranked
                    eg: Race 
                c] Interval scale data : the data can be categorized, ranked, and evenly spaced
                    eg: Temprature
                d] Ratio scale data : the data can be categorized, ranked, evenly spaced, and has a natural zero.
                    eg: Students marks in class
